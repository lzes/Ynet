```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb

%matplotlib inline
```

```python
DATA_PATH = '../test/scene/plat.pkl'
df = pd.read_pickle(DATA_PATH)
df.head()
```

```python
trajectory = df[['X', 'Y']].to_numpy().astype('float32').reshape(-1, 20, 2)
trajectory.shape
```

```python
img_path = '../test/scene/plat/reference.jpg'
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image)

colors_rgb = [
        (1.0, 0.0, 0.0),  # Red
        (0.0, 1.0, 0.0),  # Green
        (0.0, 0.0, 1.0),  # Blue
        (1.0, 1.0, 0.0),  # Yellow
        (0.5, 0.0, 0.5)   # Purple
    ]

# 为每个轨迹绘制线条，并使用不同的颜色
for i, traj in enumerate(trajectory):
    if i <= 10:
        x, y = zip(*traj)  # 分离 x 和 y 坐标
        ax.plot(x, y, marker='.', linestyle=':', color=colors_rgb[i % len(colors_rgb)], linewidth=1)
plt.show()
```
