```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def visualize_mask_and_image(image_path, mask_path, cmap_name='tab10'):
    """
    将原始图像与其对应的分割 mask 一起展示。
    
    参数:
    image_path (str): 原始图像的路径。
    mask_path (str): 分割 mask 的路径。
    cmap_name (str): Matplotlib colormap 名称，默认使用 'tab10'，可以更改以使用不同的配色方案。
    
    返回:
    None
    """
    # 读取原始图像
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # 将 BGR 转换为 RGB
    
    # 读取分割 mask
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    print(mask.min(), mask.max())
    # 创建一个颜色映射
    cmap = plt.get_cmap(cmap_name)
    
    # 将 mask 转换为 RGB 图像
    rgba_img = cmap(mask.astype(int))
    
    # 移除透明通道
    rgb_mask = np.delete(rgba_img, 3, 2)
    
    # 显示图像
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    
    # 显示原始图像
    axs[0].imshow(image)
    axs[0].set_title('Original Image')
    axs[0].axis('off')
    
    # 显示分割 mask
    axs[1].imshow(rgb_mask)
    axs[1].set_title('Segmentation Mask')
    axs[1].axis('off')
    
    plt.tight_layout()
    plt.show()
    
# 示例图像路径
image_path = 'ynet_additional_files/data/SDD/train/bookstore_0/reference.jpg'
mask_path = 'ynet_additional_files/data/SDD_semantic_maps/train_masks/bookstore_0_mask.png'

# 可视化图像和 mask
visualize_mask_and_image(image_path, mask_path)
```
